```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### choose 6 distributions from [this list](https://docs.scipy.org/doc/numpy/reference/routines.random.html#distributions)

and follow the 5 or 6 steps

```python
# 01 - gamma(shape[, scale, size])
# represents the time until an event (such as lifespan until death), when the event starts 
# out unlikely (few people die in youth), becomes more likely (more people die in old age),

gammavar = np.random.gamma(5,3, 100)

# Plot the two histograms next to each other
plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
plt.hist(gammavar)

plt.subplot(1,2,2)
plt.hist(gammavar, bins=20, color='c')

# Add a vertical line at the mean.
plt.axvline(gammavar.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at the standard deviation above the mean.
plt.axvline(gammavar.std(), color='b', linestyle='dashed', linewidth=2)

# Print the histogram.
plt.show()
```

```python
# 02 - gemoetric distribution
# The geometric distribution models the number 
# of trials that must be run in order to achieve success.

p = 0.16

geovar = np.random.geometric(p, 100)

# Plot the two histograms next to each other
plt.figure(figsize=(10, 4))

plt.subplot(1,2,1)
plt.hist(geovar, bins = 5)

plt.subplot(1,2,2)
plt.hist(geovar, bins=5, color='c')

# Add a vertical line at the mean.
plt.axvline(geovar.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at the standard deviation above the mean.
plt.axvline(geovar.std(), color='b', linestyle='dashed', linewidth=2)

# Print the histogram.
plt.show()
```

```python
# 03 - poisson
# represents the number of times a given event (such as a phone call to a radio show) 
# will occur during a given time interval, based on a count of the number of times an 
# event occurs during a given time period.


poissonvar = np.random.poisson(4, 100)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(poissonvar, color='purple')

plt.subplot(1,2,2)
plt.hist(poissonvar, color='orange')

# Add a vertical line at the mean.
plt.axvline(poissonvar.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at the standard deviation above the mean.
plt.axvline(poissonvar.std(), color='b', linestyle='dashed', linewidth=2)

# Print the histogram.
plt.show()
```

```python
# 04 - laplace
# Draw samples from the Laplace or double exponential distribution with specified 
# location (or mean) and scale (decay).

# The Laplace distribution is similar to the Gaussian/normal distribution, but is sharper 
# at the peak and has fatter tails. It represents the difference between two independent, 
# identically distributed exponential random variables.

laplacevar = np.random.laplace(0,1, 100)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(laplacevar, color='green')

plt.subplot(1,2,2)
plt.hist(laplacevar, color='pink')

# Add a vertical line at the mean.
plt.axvline(laplacevar.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at the standard deviation above the mean.
plt.axvline(laplacevar.std(), color='b', linestyle='dashed', linewidth=2)

# Print the histogram.
plt.show()
```

```python
# 05 - dirichlet
# a multivariate generalization of the beta distribution

alpha = [3,2,1]
chletvar = np.random.dirichlet(alpha, 100)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(chletvar, bins = 3)

plt.subplot(1,2,2)
plt.hist(chletvar, bins = 3)

# Add a vertical line at the mean.
plt.axvline(chletvar.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at the standard deviation above the mean.
plt.axvline(chletvar.std(), color='b', linestyle='dashed', linewidth=2)

# Print the histogram.
plt.show()
```

```python
#06 - beta

a = 1
b = 3
betavar = np.random.beta(a, b, 100)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(betavar, bins = 7)

plt.subplot(1,2,2)
plt.hist(betavar, bins = 7, color='green')

# Add a vertical line at the mean.
plt.axvline(betavar.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at the standard deviation above the mean.
plt.axvline(betavar.std(), color='b', linestyle='dashed', linewidth=2)

# Print the histogram.
plt.show()
```

### Part 2

* Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
* Add them together to create a third variable.
* Graph the third variable using a histogram.
* Compute the mean and standard deviation and plot them as vertical lines on the histogram.
* Evaluate the descriptive statistics against the data.

```python
# syntax:
# np.random.normal(mean, sd, n)

first = np.random.normal(5, 1, 100)
second = np.random.normal(10, 1,100)

third = first + second

# graph the 3rd variable
#plt.hist(third, bins=7,color='teal')

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(third, bins=5, color='teal')

plt.subplot(1,2,2)
plt.hist(third, bins=5, color='orange')

# Add a vertical line at the mean.
plt.axvline(third.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at the standard deviation above the mean.
plt.axvline(third.std(), color='b', linestyle='dashed', linewidth=2)

# Print the histogram.
plt.show()
```
